# <center> Plot Training Loss evolution for Mnist,Fashion,Cifar10
    
This notebook loads the different results with SGD and Musketeer then display the associated loss functions.

### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Tool function to load results

```python
def load_losses(name='mnist'):
    loss_sgd  = np.load('results/'+name+'/losses_'+name+'_sgd.npy')
    loss_eta1 = np.load('results/'+name+'/losses_'+name+'_e01.npy')
    loss_eta2 = np.load('results/'+name+'/losses_'+name+'_e02.npy')
    loss_eta10= np.load('results/'+name+'/losses_'+name+'_e10.npy')
    return loss_sgd,loss_eta1,loss_eta2,loss_eta10
```

### Tool function to compute moving average

```python
def movingaverage (values, window):
    weights = np.repeat(1.0, window)/window
    sma = np.convolve(values, weights, 'valid')
    return sma
```

## PLOT MNIST

```python
# Load results for MNIST
sgd,e01,e02,e10 = load_losses(name='mnist')
# Compute mean over different seeds
loss_sgd = np.mean(sgd,axis=0)
loss_e01 = np.mean(e01,axis=0)
loss_e02 = np.mean(e02,axis=0)
loss_e10 = np.mean(e10,axis=0)
```

```python
plt.figure(figsize=(5,5))
plt.plot(movingaverage(loss_sgd,10),label = 'sgd')
plt.plot(movingaverage(loss_e01,10),label = r'$\eta$=1')
plt.plot(movingaverage(loss_e02,10),label = r'$\eta$=2')
plt.plot(movingaverage(loss_e10,10),label = r'$\eta$=10')
plt.legend(fontsize=15)
plt.xlabel('Normalized passes over data (x1e3)',fontsize=15)
plt.ylabel('Training Loss',fontsize=15)
plt.xticks(ticks=np.array([0,1,2,3,4,5,6,7])*1e3,labels=[0,1,2,3,4,5,6,7],fontsize=15)
plt.yticks(fontsize=15)
plt.grid(linestyle='--',alpha=0.5)
#plt.savefig('mnist.pdf', bbox_inches='tight',transparent=True, pad_inches=0)
plt.show()
```
