# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load results

```python
def load_data(name):
    losses_uni = np.load('results/zo_'+name+'/losses_zo_'+name+'_uni.npy')
    losses_nes = np.load('results/zo_'+name+'/losses_zo_'+name+'_nes.npy')
    losses_avg = np.load('results/zo_'+name+'/losses_zo_'+name+'_avg.npy')
    losses_sqr = np.load('results/zo_'+name+'/losses_zo_'+name+'_sqr.npy')
    losses_abs = np.load('results/zo_'+name+'/losses_zo_'+name+'_abs.npy')
    return losses_uni,losses_nes,losses_avg,losses_sqr,losses_abs
```

# MNIST

```python
l_uni,l_nes,l_avg,l_sqr,l_abs = load_data('mnist')
```

```python
mean_uni = np.mean(l_uni,axis=0)
std_uni = np.std(l_uni,axis=0)

mean_nes = np.mean(l_nes,axis=0)
std_nes = np.std(l_nes,axis=0)

mean_avg = np.mean(l_avg,axis=0)
std_avg = np.std(l_avg,axis=0)

mean_sqr = np.mean(l_sqr,axis=0)
std_sqr = np.std(l_sqr,axis=0)

mean_abs = np.mean(l_abs,axis=0)
std_abs = np.std(l_abs,axis=0)
```

```python
fig,ax = plt.subplots(figsize=(5,5))
plt.plot(mean_uni,color='k',linestyle='dashdot',label='uniform')
plt.plot(mean_nes,color='purple',marker='p',markevery=0.1,ms=4,label='nesterov')
plt.plot(mean_sqr,color='red',marker='v',markevery=0.1,ms=4,label='mus_sqr')
plt.plot(mean_abs,color='green',marker='o',markevery=0.1,ms=4,label='mus_abs')
plt.plot(mean_avg,color='b',marker='s',markevery=0.1,ms=4,label='mus_avg')

plt.fill_between(np.arange(100),
                 mean_uni-std_uni/2,
                 mean_uni+std_uni/2,
                 color='k',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_nes-std_nes/2,
                 mean_nes+std_nes/2,
                 color='purple',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_avg-std_avg/2,
                 mean_avg+std_avg/2,
                 color='b',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_sqr-std_sqr/2,
                 mean_sqr+std_sqr/2,
                 color='red',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_abs-std_abs/2,
                 mean_abs+std_abs/2,
                 color='green',alpha=0.2)

plt.ylabel(r'Training Loss',fontsize=15)
plt.xlabel('#Queries of Loss f (x1e3)',fontsize=15)
plt.xticks(ticks=[0,20,40,60,80,100],labels=['0','75','150','225','300','375'],fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='upper right',fontsize=13)
plt.grid(linestyle='--',which='both',alpha=0.5)
plt.tight_layout()
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
filename = 'loss_mnist.pdf'
#plt.savefig(filename,bbox_inches='tight',transparent=True, pad_inches=0)
plt.show()
```
