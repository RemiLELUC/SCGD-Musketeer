# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load results

```python
def load_data(name):
    acc_uni = np.load('results/zo_'+name+'/acc_zo_'+name+'_uni.npy')
    acc_nes = np.load('results/zo_'+name+'/acc_zo_'+name+'_nes.npy')
    acc_avg = np.load('results/zo_'+name+'/acc_zo_'+name+'_avg.npy')
    acc_sqr = np.load('results/zo_'+name+'/acc_zo_'+name+'_sqr.npy')
    acc_abs = np.load('results/zo_'+name+'/acc_zo_'+name+'_abs.npy')
    return acc_uni,acc_nes,acc_avg,acc_sqr,acc_abs
```

# MNIST

```python
 acc_uni,acc_nes,acc_avg,acc_sqr,acc_abs = load_data('mnist')
```

```python
mean_uni = np.mean(acc_uni,axis=0)
std_uni = np.std(acc_uni,axis=0)

mean_nes = np.mean(acc_nes,axis=0)
std_nes = np.std(acc_nes,axis=0)

mean_avg = np.mean(acc_avg,axis=0)
std_avg = np.std(acc_avg,axis=0)

mean_sqr = np.mean(acc_sqr,axis=0)
std_sqr = np.std(acc_sqr,axis=0)

mean_abs = np.mean(acc_abs,axis=0)
std_abs = np.std(acc_abs,axis=0)
```

```python
print('uni:',mean_uni[-1])
print('nes:',mean_nes[-1])
print('sqr:',mean_sqr[-1])
print('avg:',mean_avg[-1])
print('abs:',mean_abs[-1])
```

```python
fig,ax = plt.subplots(figsize=(5,5))
plt.plot(mean_uni,color='k',linestyle='dashdot',label='uniform')
plt.plot(mean_nes,color='purple',marker='p',markevery=0.1,ms=4,label='nesterov')
plt.plot(mean_avg,color='b',marker='s',markevery=0.1,ms=4,label='mus_avg')
plt.plot(mean_sqr,color='red',marker='v',markevery=0.1,ms=4,label='mus_sqr')
plt.plot(mean_abs,color='green',marker='o',markevery=0.1,ms=4,label='mus_abs')

plt.fill_between(np.arange(100),
                 mean_uni-std_uni/2,
                 mean_uni+std_uni/2,
                 color='k',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_nes-std_nes/2,
                 mean_nes+std_nes/2,
                 color='purple',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_avg-std_avg/2,
                 mean_avg+std_avg/2,
                 color='b',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_sqr-std_sqr/2,
                 mean_sqr+std_sqr/2,
                 color='red',alpha=0.2)

plt.fill_between(np.arange(100),
                 mean_abs-std_abs/2,
                 mean_abs+std_abs/2,
                 color='green',alpha=0.2)

plt.ylabel(r'Test Accuracy',fontsize=15)
plt.xlabel('#Queries of Loss f (x1e3)',fontsize=15)
plt.xticks(ticks=[0,20,40,60,80,100],labels=['0','75','150','225','300','375'],fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='lower right',fontsize=15)
plt.grid(linestyle='--',which='both',alpha=0.5)
plt.tight_layout()
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
filename = 'acc_mnist.pdf'
#plt.savefig(filename,bbox_inches='tight',transparent=True, pad_inches=0)
plt.show()
```
