$f(x)=x^T Q x + b^T x$ with $Q_{ii} = b_i = 1$ if $i \in S$ and $0$ otherwise

```python
import numpy as np
from source import optimize
import matplotlib.pyplot as plt
```

```python
d = 100
s = 30
S = np.random.choice(a=d,size=s,replace=False)
b = np.zeros(d)
b[S]=1
Q = np.diag(b)
```

```python
def f_s(x):
    return x.T@Q@x + b.T@x
```

```python
gamma=1
mu_power=1/2
a=1
alpha_power=1
N=10
T=75
eta=1
x0 = np.zeros(d)
```

```python
_,x_full = optimize(seed=0,d=d,f=f_s,gamma=gamma,
                    mu_power=mu_power,x0=x0,N=N,
                    T=T,a=a,alpha_power=alpha_power,eta=None,
                    method='full',unbiased=None,verbose=False)
```

```python
_,x_uni = optimize(seed=0,d=d,f=f_s,gamma=gamma,
                    mu_power=mu_power,x0=x0,N=N,
                    T=T,a=a,alpha_power=alpha_power,eta=None,
                    method='uni',unbiased=None,verbose=False)
```

```python
_,x_bia = optimize(seed=0,d=d,f=f_s,gamma=gamma,
                    mu_power=mu_power,x0=x0,N=N,
                    T=T,a=a,alpha_power=alpha_power,eta=eta,
                    method='musketeer',unbiased=False,verbose=True)
```

```python
_,x_unb = optimize(seed=0,d=d,f=f_s,gamma=gamma,
                    mu_power=mu_power,x0=x0,N=N,
                    T=T,a=a,alpha_power=alpha_power,eta=eta,
                    method='musketeer',unbiased=True,verbose=False)
```

```python
loss_ful = np.array([f_s(x) for x in x_full])
loss_uni = np.array([f_s(x) for x in x_uni])
loss_bia = np.array([f_s(x) for x in x_unb])
loss_unb = np.array([f_s(x) for x in x_bia])
```

```python
reg_ful = np.cumsum(loss_ful)/np.arange(1,N+2)
reg_uni = np.cumsum(loss_uni)/np.arange(1,N+2)
reg_bia = np.cumsum(loss_bia)/np.arange(1,N+2)
reg_unb = np.cumsum(loss_unb)/np.arange(1,N+2)
```

### Plot results

```python
plt.figure(figsize=(10,5))
#plt.yscale('symlog')
plt.plot(loss_ful,label='ful')
plt.plot(loss_uni,label='uni')
plt.plot(loss_bia,label='bia')
plt.plot(loss_unb,label='unb')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(reg_ful,label='ful')
plt.plot(reg_uni,label='uni')
plt.plot(reg_bia,label='bia')
plt.plot(reg_unb,label='unb')
plt.legend()
plt.show()
```
